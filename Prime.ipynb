```python
import pandas as pd
import ast
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TRADES_CopyTr_90D_ROI.csv')
```

```python

df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Convert 'Trade_History' column entries from string representation of lists to actual lists
df['Trade_History'] = df['Trade_History'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

```

```python
all_trades = []
```

```python
for index in df.index:
    port_id = df.at[index, 'Port_IDs'] # Get the Port_ID for the current index
    trades = df.at[index, 'Trade_History']  # Get the trade history for the current index
    
    
    if isinstance(trades, list):
        for trade in trades:
           
            trade_with_port_id = trade.copy()
            trade_with_port_id['Port_ID'] = port_id  
            all_trades.append(trade_with_port_id)  
    else:
        print(f"Skipping non-list trade entry for index {index}: {trades}")
```

```python
df_trades = pd.DataFrame(all_trades)
```

```python
df_trade = pd.read_csv('Final_trade.csv')
```

```python
df_trade.head()
```

```python
df_trade.info()
```

```python
# Create a new column to represent the position type based on 'side' and 'positionSide'
df_trade['Position_Type'] = df_trade.apply(lambda x: f"{x['side']}_{x['positionSide']}", axis = 1)
```

```python
# Calculate the trade value by multiplying quantity by price
df_trade['Trade_value'] = df_trade['quantity'] * df_trade['price']
```

```python
df_trade.head()
```

```python
# Aggregate metrics for each Port_ID
Metrics = df_trade.groupby('Port_ID').agg(
    pnl = ('realizedProfit','sum'),
    Win_Positions = ('realizedProfit', lambda x: (x > 0).sum()),
    Total_Positions = ('realizedProfit', 'count'),
).reset_index()

```

```python
# Calculate ROI
Metrics['ROI'] = (Metrics["pnl"] / (Metrics['pnl'].abs().sum()))* 100
```

```python
# Calculate winrate
Metrics['Win_rate'] = (Metrics["Win_Positions"] / Metrics['Total_Positions'])* 100
```

```python
# Calculate return for each trade as realized profit divided by quantity
df_trade['Return'] = df_trade['realizedProfit']/ df_trade['quantity']
```

```python
df_trade.head()
```

```python

# Calculate Sharpe Ratio for each Port_ID
sharpe_ratio = df_trade.groupby('Port_ID')['Return'].agg(
    mean_return=lambda x: x.mean(),  
    std_return=lambda x: x.std()  
)
```

```python
risk_free_rate = 0.01
```

```python
sharpe_ratio['Sharpe_Ratio'] = (sharpe_ratio['mean_return'] - risk_free_rate) / sharpe_ratio['std_return']
```

```python
Metrics = Metrics.merge(sharpe_ratio[['Sharpe_Ratio']], on='Port_ID', how='left')
```

```python
# Function to calculate the maximum drawdown from returns
def max_drawdown(returns):
    cumulative_returns = (1 + returns).cumprod()
    peak = cumulative_returns.cummax()
    drawdown = (cumulative_returns - peak) / peak
    return drawdown.min()
# Calculate maximum drawdown for each Port_ID
mdd = df_trade.groupby('Port_ID')['Return'].agg(max_drawdown)
Metrics = Metrics.merge(mdd.rename('MDD'), on='Port_ID')
```

```python
# Create a score based on weighted contributions of various metrics
Metrics['Score'] = (Metrics['pnl']* 0.5) + (Metrics['Win_Positions'] * 0.3) + (Metrics['ROI'] * 0.2)
```

```python
# Rank accounts based on the calculated score
Metrics['Rank'] = Metrics['Score'].rank(ascending=False)
```

```python
Metrics.to_csv('Final_account.csv', index=False)
```
